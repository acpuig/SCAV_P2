## Exercise 1.

**Create a script which can cut N seconds
the BBB video. That N seconds will be read
from user input.**

Per aquest exercici utilitzo dos parametres, el més important és _-ss_ que ens extraura una porcio de video, aquesta marcara l'inici del tall. Després el paramtre _-to_ ens servirà per especificar el temps fins el final de tall. 


```python
N= 30 #seconds
ff = ffmpy.FFmpeg(
        inputs={'input.mp4': None},
        outputs={'EX1_output.mp4': '-ss 00:00:00 -to 00:00:'+str(N)+''}
    )
ff.cmd
ff.run()
```

## Exercise 2.

**Create a script which will extract the YUV
histogram from the previous BBB video you’ve
done and create a new video with both images
at the same time.**

Per aquest exercici utilitzo el parametre _-vf_. He tret la opció de fer la superposició al video de: https://trac.ffmpeg.org/wiki/Histogram

```python
ff = ffmpy.FFmpeg(
        inputs={'EX1_output.mp4': None},
        outputs={'EX2_output.mp4': '-vf "split=2[a][b],[b]histogram,format=yuva444p[hh],[a][hh]overlay"'}
    )
ff.cmd
ff.run()
```

## Exercise 3.

**Create a script which can resize the BBB
video into 4 differents video outputs (doesn’t
need to be at the same time):**

        ·720p (1280x720) ·480p (720x480) ·360x240 ·160x120

Per fer l'exercici utilitzo el parametre _scale_ que realitzara una escalada de la qualitat del video. 

```python
ff = ffmpy.FFmpeg(
        inputs={'EX1_output.mp4': None},
        outputs={'EX3_720.mp4': '-vf scale=1280:720'}
    )
ff.cmd
ff.run()
ff = ffmpy.FFmpeg(
        inputs={'EX1_output.mp4': None},
        outputs={'EX3_489.mp4': '-vf scale=720:480'}
    )
ff.cmd
ff.run()
ff = ffmpy.FFmpeg(
        inputs={'EX1_output.mp4': None},
        outputs={'EX3_360x240.mp4': '-vf scale=360:240'}
    )
ff.cmd
ff.run()
ff = ffmpy.FFmpeg(
        inputs={'EX1_output.mp4': None},
        outputs={'EX3_160x120.mp4': '-vf scale=160:120'}
    )
ff.cmd
ff.run()
```

## Exercise 4.

**Create a script which can change the audio
into mono output and in a different audio codec**

Aquest exercici constara de dues parts. Primer passarem el video a mono utilitzant _-ac_ trobat a https://www.codegrepper.com/code-examples/whatever/stereo+to+mono+ffmpeg. Després utilitzem _-acodec_ trobat a la pàgina de _FFmpeg_. 

```python
ff = ffmpy.FFmpeg(
        inputs={'EX1_output.mp4': None},
        outputs={'EX4_mono.mp4': '-ac 1'}
    )
ff.cmd
ff.run()
ff2 = ffmpy.FFmpeg(
        inputs={'EX4_mono.mp4': None},
        outputs={'EX4_codec_mp3.mp4': '-acodec mp3'}
    )
ff2.cmd
ff2.run()
```
